```python
import spacy
```

```python
core_esp = spacy.load("es_core_news_sm")
```

```python
Texto = core_esp(u"Cuando cuentes cuentos cuenta cuantos cuentos cuentas, así, sabrás cuantos cuentos contaste")

contPalabras = 0
contLemma = []
band = False

for token in Texto:
    contPalabras += 1
    band = False
    print("Texto:", token.text) #.text extrae el texto del token
    print("PoS:", token.pos_) #.pos_ extrae el part of speech
    print("Lemma:", token.lemma_) #.lemma_ extrae la lematización
    for i in contLemma:
      if i == token.lemma:
          band = True

    if band == False:
        contLemma.append(token.lemma)
    Dificultad = len(contLemma)/contPalabras
    print("\n")

print(Dificultad)


```
